```python
 # 01: Preprocessing imports
import pandas as pd
import numpy as np
```

# 2. Importing Data
<hr style="border-top:1px dashed">

```python
#Creating empty lists to import each csv file into
DP_tables = []
Gwent=[]
North_wales=[]
South_wales=[]


#running for loops for data spanning 2 years and appending each to lists created above
for i in ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']:
    for j in ['2023', '2024']:
        table = pd.read_csv(f'Data/Dyfed_Powys/{j}-{i}-Dyfed-Powys-street.csv')
        DP_tables.append(table)

for i in ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']:
    for j in ['2023', '2024']:
        table = pd.read_csv(f'Data/Gwent/{j}-{i}-Gwent-street.csv')
        DP_tables.append(table)

for i in ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']:
    for j in ['2023', '2024']:
        table = pd.read_csv(f'Data/North_wales/{j}-{i}-north-wales-street.csv')
        DP_tables.append(table)

for i in ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']:
    for j in ['2023', '2024']:
        table = pd.read_csv(f'Data/South_wales/{j}-{i}-south-wales-street.csv')
        DP_tables.append(table)

#combining all lists into one list
all_crime = DP_tables + Gwent + North_wales + South_wales
```

```python
#Changing the format from list to dataframe
final_crime = pd.concat(all_crime)

original_crime=final_crime.copy()

original_crime
```

<hr style="border:2px solid">

# 3. Data Cleaning
<hr style="border-top:1px dashed">

```python
#Check data shape

final_crime.shape
```

```python
#Checking for nulls

final_crime.isnull().sum()
```

```python
#Checking data types

final_crime.dtypes
```

```python
#Checking if falls within and reported by are the same

print((final_crime['Reported by'] != final_crime['Falls within']).sum())
```

**Deleting Columns**
  - Context column has been dropped as it is empty
  - Based on the test above falls within has been dropped
  - LSOA code has been dropped as we already have the corresponding area
  - Crime ID has been dropped as there are many missing and it doesn't offer valuable information

```python
# Deleting useless columns


final_crime.drop(columns=['Context','LSOA code', 'Crime ID', 'Falls within'], inplace=True)
```

```python
# Filling in unknown values and dropping if only small amounts of data is lost

final_crime['Last outcome category']=final_crime['Last outcome category'].fillna('Unknown')

final_crime['LSOA name']=final_crime['LSOA name'].fillna('Unknown')

final_crime = final_crime.dropna()
```

```python
# Changing Data Formats and separating out the month from the year

final_crime['Month'] = pd.to_datetime(final_crime['Month'], format='%Y-%m')

final_crime['year'] = final_crime['Month'].dt.year
final_crime['month'] = final_crime['Month'].dt.month

final_crime.drop(columns= 'Month', inplace=True)


final_crime
```

```python
#Checking values within columns (run each individually for results of each check)

final_crime['Reported by'].unique()

final_crime['Crime type'].unique()

final_crime['Last outcome category'].unique()
```

```python
# Standardising format of columns that have too many different inputs to manually check like above

final_crime['Location'] = final_crime['Location'].str.title()

final_crime['LSOA name'] = final_crime['LSOA name'].str.title()

final_crime
```
